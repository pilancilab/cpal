```python
import glob
import pickle
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Directory where your pkl files are stored
directory_path = 'spiral_data'  # replace this with your actual directory

# Glob pattern to match your files
file_pattern = os.path.join(directory_path, 'acc_*_*.pkl')

# Dictionary to store the loaded datasets
datasets = {}

# Function to parse file name and load data
def load_and_parse_files(file_pattern):
    for filepath in glob.glob(file_pattern):
        # Extract parts from the filename
        filename = os.path.basename(filepath)
        parts = filename.split('_')
        
        strategy = parts[1]  # Strategy name
        data_type = parts[2]  # Train or test
        seed = parts[3]  # Seed (e.g., seed4)
        
        # Handle cases where there is no '_complete'
        if len(parts) == 5:
            # There is a '_complete'
            complete = parts[4].split('.')[0]  # 'complete' without the '.pkl'
        else:
            # No '_complete'
            complete = None
        
        # Load the dataset
        with open(filepath, 'rb') as file:
            data = pickle.load(file)
        
        # Store the data in a dictionary using the parsed info
        # Include 'complete' as part of the key if it exists
        key = (strategy, data_type, seed, complete) if complete else (strategy, data_type, seed)
        datasets[key] = data
```

```python
# Call the function to load the datasets
load_and_parse_files(file_pattern)

# Now you can access the datasets, e.g.,
# datasets[('RandomSampling', 'train50', 'seed4')]
# datasets[('RandomSampling', 'train50', 'seed4', 'complete')] if complete exists
```

```python
len(datasets[('LeastConfidence', 'train50', 'seed0', 'complete')])
```

```python
itr_ls = list(range(0,55,5))
itr_ls[0] = 1
itr_ls = list(np.array(itr_ls)-1)
itr_ls
```

```python
for key, value in datasets.items():
    # process complete
    if 'complete' in key:
        datasets[key] = [value[i] for i in itr_ls if i < len(value)]
```

```python
for i, (key, value) in enumerate(datasets.items()):
    if i < 5:
        print(f"Key: {key}")
        print(f"Value: {value}\n")  # Adjust this if you want to print only part of the value
    else:
        break
```

```python
# The order of strategies based on the image
desired_order = [
    'greedyX', 
    'EntropySampling', 
    'RandomSampling', 
    'LeastConfidence', 
    'BALDDropout', 
    'qbc', 
    'linear', 
    'cvx'  # cvx corresponds to 'Cutting-Plane (ours)'
]

# Initialize dictionary to store results
strategy_results = {}

# Iterate through the datasets to collect test results across seeds
for key, value in datasets.items():
    if 'complete' in key:
        strategy, data_type, seed, complete = key
    else:
        strategy, data_type, seed = key
    
    if data_type.startswith('test'):  # Focus on test data
        # Initialize list for each strategy if not already done
        if strategy not in strategy_results:
            strategy_results[strategy] = []
        # Append the value (test result) for this seed
        strategy_results[strategy].append(value)

# Reorder the dictionary according to the desired order
strategy_results = {strategy: strategy_results[strategy] for strategy in desired_order if strategy in strategy_results}
```

```python
# Prepare lists for plotting
strategies = []
means = []
std_devs = []

# Compute mean and standard deviation for each strategy
for strategy, values in strategy_results.items():
    strategies.append(strategy)
    means.append(np.mean(np.array(values), axis = 0))  # Mean performance across seeds
    std_devs.append(np.std(np.array(values), axis = 0))  # Standard deviation across seeds
```

```python
num_queries = np.array(itr_ls)+1

# Plotting the error bar chart
plt.figure(figsize=(10, 6))
for i in range(len(means)):
    if i == len(means)-1:
        #plt.semilogy(num_queries, means[i], color = 'red', label = 'f{strategy}', marker = 'o') 
        plt.errorbar(num_queries, means[i], color = 'red', label = 'Cutting-Plane (ours)', yerr=std_devs[i], marker = 'o')
    else:
        #plt.semilogy(num_queries, means[i], label = 'f{strategy}', marker = 'x')
        plt.errorbar(num_queries, means[i], yerr=std_devs[i], label = f'{strategies[i]}', marker = 'x')
#plt.yscale('log') 
#plt.xscale('linear')
plt.title('Train Accuracy Mean versus Number of Queries on Binary Spiral')
plt.xlabel('Number of Queries')
plt.ylabel('Train Accuracy Mean')
#plt.xticks(rotation=45)  # Rotate strategy labels if needed
plt.grid(True)
plt.legend()
# Show the plot
plt.tight_layout()
plt.savefig('spiralTrain_error_bar.pdf', bbox_inches='tight')
plt.show()
```
